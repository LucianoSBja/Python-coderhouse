Los números se pasan por valor y crean una copia dentro de la función, no les afecta externamente lo que hagamos con ellos en la función:


```python
def doblar_valor(numero):
    numero = numero * 2
    return numero
```

```python
lista_numeros = [10, 50, 100]
```

```python
def obtener_dobles(numeros):
    for indice in range(len(numeros)):
        # Obtengo el numero actual
        numero_actual = numeros[indice]

        # Actualizo el elemento en la lista con su doble
        numeros[indice] = numero_actual*2
    return numeros
```

```python
obtener_dobles(lista_numeros)
```

```python
lista_numeros
```

```python
# truco para pasar listas o diccionarios por valor
```

```python
# Yo no quiero que la lista sea actualizada despúes de correr obtener_dobles
lista_2 = [30, 40, 50]
```

```python
# El truco es enviar una copia de la lista original
# Opcion 1
obtener_dobles(lista_2[:])
```

```python
# Opcion 2
obtener_dobles(lista_2.copy())
```

```python
lista_2
```

# Funciones recursivas
Son funciones que dentro de la definición se llaman así mismas

```python
# Ejemplo factorial
# pregunta reto selección a trabajos

# factorial de 5 es igual a 5*4!

def factorial(num):
    if num == 1:
        return 1
    else:
        return num * factorial(num - 1)
```

```python
factorial(1)
```

```python
factorial(5)
```

```python
5*factorial(4)
```

```python
def cuenta_regresiva(n):
    print(f"Funcion llamada con argumento {n}")
    if n == 0:
        print("¡Despegue!")
    else:
        print(f"inicio del bloque: {n}")
        cuenta_regresiva(n - 1)
        # Esta linea no se va a imprimir hasta que termine la llamada recursiva de la linea de arriba
        print(f"fin del bloque: {n}")
```

```python
cuenta_regresiva(5)
```
