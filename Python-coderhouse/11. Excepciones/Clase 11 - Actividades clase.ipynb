```python
"""
En la función de nuestro ejercicio hay un fallo potencial, averigua cuándo sucede y
retorna el valor None en ese caso especial, en cualquier otro caso devuelve el resultado
normal de la función.
"""

def dividir(a, b):
    return a/b
```

```python
def dividir(a, b):
    if b == 0:
        return None
    else:
        return a/b
```

```python
dividir(10,5)
```

```python
retorno = dividir(10,0)
print(retorno)
```

```python
def dividir(a, b):
    try:
        return a/b
    except ZeroDivisionError as e:
        print("No se puede dividir entre cero")
        raise
```

```python
dividir(10,5)
```

```python
dividir(10,0)
```
