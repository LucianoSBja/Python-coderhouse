### Primeras funciones

````python
# es un funcion porque tiene return explicito

def saludar_a_persona(nombre):
    saludo = f"Hola {nombre}, ¿Qué onda?"
    return saludo
````

````python
# Así usan la función, llamandola por su nombre
# llamar la función
saludar_a_persona("Ana")
````

````python
saludar_a_persona("Ender")
````

````python
alumno = input("Cual es tu nombre? ")
saludar_a_persona(alumno)
````

````python
# Así se guarda el retorno de la llamada de una función
retorno_de_saludo = saludar_a_persona("Sofia")
````

````python
retorno_de_saludo
````

````python
type(retorno_de_saludo)
````

````python
# es un procedimiento, porque al final no tiene la sentencia return. No tiene return explicito

def saludar(persona):
    saludo = f"Hola {persona}, ¿Qué onda?"
    print(saludo)
````

````python
saludar("Rafael")
````

````python
retorno_del_procedimiento = saludar("Rafael")
````

````python
type(retorno_del_procedimiento)
````

````python
retorno_del_procedimiento
````

````python
retorno_del_procedimiento is None
````

````python
# aca defini la funcion
# "estas creando la herramienta"

def sumar(numero1, numero2):
    return numero1 + numero2
````

````python
# esta es la llamada de la funcion
# es cuando uso la funcion
# estas "usando la herramienta"

sumar(10, 15)
````

### Ejemplos vida real

````python
"""
En todos los sistemas web se necesita el nombre completo del usuario, pero en la base de datos guardamos
el nombre y apellido por separado
Considera los datos de nombres y apellidos pueden tener formatos incorrectos.
Se desea la respuesta siempre de el mismo formato:
Apellido, Nombre

Ejemplo data
Es comun almacenar la data de un cliente en un diccionario, por ejemplo:
data = {
    "nombre": "MANUEL",
    "apellido": "Gonzales",
    "ciudad": "Medellin",
    "dni": "AB550000"
}
"""

def obtener_nombre_completo(diccionario):
    respuesta = f'{diccionario["apellido"]}, {diccionario["nombre"]}'
    return respuesta.title()
````

````python
data = {
    "nombre": "IVAN",
    "apellido": "tomasevich",
    "ciudad": "Medellin",
    "dni": "AB550000"
}
````

````python
obtener_nombre_completo(data)
````

## Comentarios en funciones
