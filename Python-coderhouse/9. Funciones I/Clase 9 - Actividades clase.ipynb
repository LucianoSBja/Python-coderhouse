```python
"""
Escribir una función que reciba una muestra de números en una lista y devuelva su media.

Media y promedio son sinonimos
Definicion matematica de media:
(suma de los numeros) / (cantidad de numeros)
"""
```

```python
def calcular_media(lista):
    suma = sum(lista)
    cantidad = len(lista)
    media = suma / cantidad
    return media
```

## Solución profesor

```python
def obtener_media(lista):
    suma = sum(lista)
    cantidad = len(lista)
    media = suma / cantidad
    return media
```

# Actividad 3

```python
"""
Actividad 3

Crea un programa que le pida dos números enteros al usuario y diga por consola, si alguno de
ellos es múltiplo del otro.

La función debe llamarse es_multiplo().
"""
```

```python
def es_multiplo(numero_1, numero_2):
    if numero_1 % numero_2 == 0:
        return "es multiplo"
    elif numero_2 % numero_1 == 0:
        return "es multiplo"
    else:
        return "no es multiplo"
```

## Solución profesor

```python
def es_multiplo(numero_1, numero_2):
    if numero_1 % numero_2 == 0 or numero_2 % numero_1 == 0:
        return True
    else:
        return False
```

```python
# Para usarla, la llamamos
es_multiplo(10, 2)
```

```python
es_multiplo(5, 15)
```

```python
es_multiplo(15, 5)
```

```python
es_multiplo(7, 5)
```
