```python
# Algunas veces los sets (conjuntos) parecen ordenados cuando los imprimes o retornas, pero un conjunto nunca est치 ordenado.
# Que los conjuntos no esten ordenados significa que la definici칩n de conjunto en Python no asegura un orden
# Python suele imprimir conjuntos "peque침os" como si estuvieran ordenados, pero NO lo est치n!!!
# Cualquier orden aparente es una particularidad del interprete de Python que estes usando, de la cual no puedes depender
```

```python
import random

# Generemos 50 numeros al azar del 1 al 1000
random_numbers = random.sample(range(1, 1001), 50)

# Convert the random numbers to a set to remove duplicates
random_numbers = set(random_numbers)

# Print the random numbers
print(random_numbers)
```

### Los sets se pueden iterar con un for

```python
#Uso de la funcion: Set()
frutas = set(("manzana", "banana", "cherry"))
print(frutas)
```

```python
#Los sets son objetos iterables
for fruta in frutas:
  print(fruta)
```

# Diccionarios

Un diccionario es una estructura de datos que almacena la informaci칩n en forma de pares con una clave 칰nica y un valor asociado. Existen diferentes maneras de crear un diccionario, veamos los casos mas generales:

```python
# Link es el personaje principal de Zelda
# cada par de elementos => llave: valor

link = {
    "nombre": "Link",
    "edad": 7,
    "vida_total": 10,  # corazones
    "vida_actual": 5,
    "color_vesturio": "verde",
    "lanza_fuego": False,
    "objetos": ["piedra", "espada", "escudo"]
}
```

```python
link
```

```python
link["vida_actual"]
```

```python
# diccionario usado para traducir colores del espa침ol al ingles


# clave en espa침ol. Valor en ingles
# colores = {"amarillo": "yellow", "azul": "blue", "rojo": "red"}
colores = {
    "amarillo": "yellow",
    "rojo": "red",
    "azul": "blue"
}
```

Para acceder al valor de un elemento, podemos realizarlo de la siguiente manera
