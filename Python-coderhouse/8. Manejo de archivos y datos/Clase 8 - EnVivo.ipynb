```python
# Deben usar una ruta de su drive que tenga permiso de escritura y lectura
ruta = '/drive/MyDrive/python_clase_8/'
```

## Crear archivos

```python
# .txt hace referencia a un archivo de texto
ruta_archivo = ruta + "archivo.txt"  # El nombre que le quieren dar al archivo

# f (file) es la variable en memoria del archivo
# open abre un archivo en memoria en un modo. "w" significa modo de escritura
f = open(ruta_archivo, "w")

# el comando write escribe una linea en el archivo
f.write("Esto es un texto. Hola comision 47780.")

# El .close cierra el archivo y lo guarda en el disco duro
f.close()
```

## Leer archivos

```python
# Leer archivos

# Ruta del archivo a leer
ruta_lectura = ruta + "paraLeer.txt"

# Abro el archivo en modo lectura
# "r" significa modo "read" de lectura
# archivo es una variable que tiene asignada el archivo en memoria
archivo = open(ruta_lectura, "r")

# archivo.read nos permite leer el contenido de un archivo
content = archivo.read()

print(content)
archivo.close()
```

```python
f = open(ruta_lectura, "r")

# .readlines obtengo una lista, donde cada elemento es una lina del texto
lines = f.readlines()

f.close()
```

## Librería JSON

Sirve para transformar diccionarios en formato JSON o viceversa

EQUIVALENCIA
En memoria RAM => diccionario de python
En archivo => un archivo JSON

#### Tiene dos comandos:
* json.dump que sirve para volcar un diccionario en un arcivo
* json.load que sirve para leer un archivo JSON y retornarlo como un diccionario de Python

## Guardar archivo JSON

```python
# Quiero guardar los datos del diccionario persona de forma persistente
# Para ello lo voy a guardar en un archivo .json

# 1er paso, importar libreria JSON.
import json

# La ruta donde guardaremos el archivo JSON
ruta_json = ruta + "persona.json"

# archivo_en_memoria es el nombre de la variable
archivo_en_memoria = open(ruta_json, 'w')

# json.dump crea archivos .json. Guarda un diccionario de python en un archivo
# SINTAXIS: json.dump(diccionario, archivo_en_memoria, indent)
# el indent es la sangria que va a tener el archivo, indent=4 significa 4 espacios
json.dump(persona, archivo_en_memoria, indent=4)

archivo_en_memoria.close()
```

```python
import json  # Importar libreria JSON.

ruta_json = ruta + "persona.json"

# El with se encarga de abrir y cerrar el archivo por nosotros
# Es decir no es necesario hacer archivo_en_memoria.close()
with open(ruta_json, 'w') as archivo_en_memoria:
    json.dump(persona, archivo_en_memoria, indent=4)
```

## Leer archivo JSON

```python
# Hacemos esto para leer el JSON y convertirlo en un diccionario
with open(ruta_json, "r") as archivo_en_memoria:
    # json.load lee un archivo .json y lo transforma en un objeto en memoria RAM
    data_lectura = json.load(archivo_en_memoria)
```
