# Repaso 2: Dada esta lista de números encontrar el valor mínimo y la ubicación.   [21,33,-25,85,-3.6, 27]

```python
# -25

lista = [21,33,-25,85,-3.6, 27]

minimo = lista[0]

for x in lista:

  if ( x < minimo):

    minimo = x


print(f"MINIMO: {minimo}")

#Tarea para la casa !!!!!!!! ---- en qué posición estaba el -25
```

### Conjuntos

```python
conj1 = {1,2,4,"Pepeto"}
conj2 = {2,6,"Pepe"}
conj3 = {1,2,4,"Pepeto",8,11}
conj4 = {"hola", "chao"}
```

```python
# No comparten ningun elemento en comun?
conj1.isdisjoint(conj2)
```

```python
conj3.isdisjoint(conj4)
```

```python
# El conjunto 1 esta contenido en el conjunto 3
conj1.issubset(conj3)  # El conjunto 1 es un subset del conjunto 3?
```

```python
conj3.issubset(conj1)
```

```python
# El conjunto 3 contiene el conjunto 1
conj3.issuperset(conj1)  # El conjunto 3 es un superset del conjunto 1
```
