```python
"""
Utilizando todo lo que sabes sobre cadenas, listas y sus métodos internos, transforma este texto:
"""

texto = "gordon lanzó su curva&strawberry ha fallado por un pie! -gritó Joe Castiglione&dos pies le corrigió Troop&strawberry menea la cabeza como disgustado… -agrega el comentarista"

# Transformar a:
# Gordon lanzó su curva...
# - Strawberry ha fallado por un pie! -gritó Joe Castiglione.
# - Dos pies le corrigió Troop.
# - Strawberry menea la cabeza como disgustado… -agrega el comentarista.

```

## Solución Profesor

```python
lineas_sin_editar = texto.split("&")
print(lineas_sin_editar)
```

```python
list(enumerate(lineas_sin_editar))
```

```python
# Solución usando for
lineas_editadas = []

# Vamos a iterar sobre una lista de tuplas
for (indice, texto) in enumerate(lineas_sin_editar):
    # la variable texto contiene una de las lineas del poema
    # primera letra en mayuscula + el resto del texto
    texto_editado = texto[0].upper() + texto[1:]

    # Chequeo el indice de la linea
    if indice == 0:
        # Indice 0
        texto_editado = texto_editado + "..."
    else:
        # Indices 1, 2 y 3
        texto_editado = "- " + texto_editado + "."

    # Guardar el texto editado
    lineas_editadas.append(texto_editado)
```

```python
lineas_editadas
```

```python
# Si imprimo linea solamente me devuelve el ultimo item de la lista, entonces tengo que hacer un FOR
# print(linea)

# Mostramos el texto final
for linea in lineas_editadas:
    print(linea)
```

```python
#Ejemplo 1:
my_set_1 = set([1, 2, 3])
my_set_2 = set([3, 4, 5])
my_new_set = my_set_1.union(my_set_2)
# la union nos retorna todos los elementos
print(my_new_set)
```

```python
#Interpretacion - ejemplo 1: Con la funcion union se unen los sets previamente definidos sin considerar el
#elemento repetido de 3.
```

```python
#Ejemplo 2:

my_set_1 = set([1, 2, 3])
my_set_2 = set([3, 4, 5])
my_new_set = my_set_1.intersection(my_set_2)
# la interseccion nos retorna solo los elementos en comun
print(my_new_set)
```

```python
#Interpretacion - ejemplo 2: Unicamente me reporta la interseccion entre ambos conjuntos, en este caso el elemento 3
```

```python
#Ejemplo 3:
my_set_1 = set([1, 2, 3])
my_set_2 = set([3, 4, 5])

# al 1 le quitas los elementos repetidos en el 2
my_new_set = my_set_1.difference(my_set_2)
print(my_new_set)
```

```python
#Interpretacion - ejemplo 3: Me muestra los elementos que difieren entre el set 1 y el set 2
# elementos exclusivos de 1
```
