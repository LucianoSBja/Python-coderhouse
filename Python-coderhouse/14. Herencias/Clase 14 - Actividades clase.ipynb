# Actividad: herencia multiple
Realiza la herencia múltiple respetando este diagrama de Clases UML. Y crear dos Cetáceos.


```python
class Mamifero:
    def __init__(self, mamas, vida):
       self.cantidad_mamas = mamas
       self.esperanza_de_vida = vida

    def mamar(self):
        return "Esta mamando"

    def nacer(self):
        return 'NACIENDO'


class AnimalMarino:
    def __init__(self, tiene_branqueas, especie):
        self.tiene_branqueas = tiene_branqueas
        self.especie = especie

    def nadar(self):
        return 'Nadando...'


class Cetaceo(Mamifero, AnimalMarino):
    def __init__(self, mamas, vida, tiene_branqueas, especie, notas, vive_en, peso):
        # llamando al super, llamo al __init__ de Mamifero
        super().__init__(mamas, vida)  # agrega atributos mamas y vida
        self.tiene_branqueas = tiene_branqueas
        self.especie = especie
        self.notas = notas
        self.vive_en = vive_en
        self.peso = peso

    def nacer(self):
        return 'Cetaceo naciendo en el agua'

    def nadar(self):
        return 'Nadando como un CETACEO'
```

## Solución profesor

```python
class Mamifero:

    def __init__(self, mamas, vida):
       self.cantidad_mamas = mamas
       self.esperanza_de_vida = vida

    def mamar(self):
        print('COMIENDO')

    def nacer(self):
        print('NACIENDO')


class AnimalMarino:

    def __init__(self, tiene_branqueas, especie):
        self.tiene_branqueas = tiene_branqueas
        self.especie = especie

    def nadar(self):
        print('NADANDO')



class Cetaceo(Mamifero, AnimalMarino):

    def __init__(self, notas, vive_en, peso, tiene_branqueas, especie, mamas, vida):

        # Cuando son herencias multiples hay que definir con que clases trabajar
        # super() funciona solo con 1

        # Mamifero
        Mamifero.__init__(self, mamas=mamas, vida=vida)  # agrega atributos mamas y vida

        # AnimalMarino
        AnimalMarino.__init__(self, tiene_branqueas=tiene_branqueas, especie=especie)  # agrega atributos tiene_branqueas y especie

        self.notas = notas
        self.vive_en = vive_en
        self.peso = peso

    # Voy a sobreescribir nacer
    def nacer(self):
        print('CETACEO NACE')

    def nadar(self):
        super().nadar()
        print('Sigue Nadando...')
```

```python
Cetaceo.__mro__
```

```python
delfin_comun = Cetaceo(
    vive_en='oceona atlantico',
    notas='Es un animal inteligente',
    peso=101,
    especie='Delfin Común',
    mamas=4,
    tiene_branqueas=False,
    vida=60)
```

```python
delfin_comun.especie # de AnimalMarino
```

```python
delfin_comun.esperanza_de_vida  # de Mamifero
```

```python
delfin_comun.peso  # de Cetaceo
```

```python
delfin_comun.nacer()
```

```python
# Fue heredado
delfin_comun.mamar()
```
